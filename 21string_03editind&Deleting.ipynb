```python
# strings are immutable in python
# trying to change a character in a string will result in an error
s = "hello"
# s[0] = 'H'  # This will raise a TypeError

# so editing is not allowed, but we can create a new string
```

```python
# we can reassign the variable to a new string
s = "Hello"
print(s)
s="Hello, World!"
print(s)
```

```python
# deletion of strings is not allowed
# del s[0]  # This will raise a TypeError   
# but we can delete the entire string variable
del s
# print(s)  # This will raise a NameError since s is deleted
print(s)
```
