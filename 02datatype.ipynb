```python
# python supports 3 types of datatypes
# 1. Primitive data types   - integer, float, boolean, string, complex
# 2. Container data types  - list, tuple, set, dictionary                        
# 3. User-defined data types - class, object
```

```python
# python can print upto 1e308 integer values
print(1e308)
```

```python
print(1e309)  # prints inf
```

```python
# float
print(3.14)
print(1.7e308)  # prints 1.7e+308
print(1.8e308)  # prints inf
print(-1.8e308)  # prints -inf
print(1.7e309)  # prints inf
print(-1.7e309)  # prints -inf
```

```python
#boolean
print(True)
print(False)
```

```python
#complex
print(2 + 3j)   
```

```python
#string
print('Hello, World!')
print("Hello, World!")
print("""Hello, World!""")
```

```python
#list
print([1, 2, 3, 4, 5])
```

```python
#tuple
print((1, 2, 3, 4, 5))
```

```python
#sets
print({1, 2, 3, 4, 5})
```

```python
#dictionary
print({'name': 'Alice', 'age': 30, 'city': 'New York'})
```
