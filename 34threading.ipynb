```python
#  using thread

from time import time,sleep
import threading

start_time=time()
def dosomething(id):

    print("Going to sleep...",id)
    sleep(1)
    print(f"Woken up...{id}")

thread1=threading.Thread(target=dosomething,args=[0])
thread1.start()

thread1.join() # main thread will wait for thread1
end_time=time()

print(f"main thread finishes at {end_time-start_time}")
```

```python
#  using thread but not waiting for child to be finished execution

from time import time,sleep
import threading

start_time=time()
def dosomething(id):

    print("Going to sleep...",id)
    sleep(1)
    print(f"Woken up...{id}")

threads=[threading.Thread(target=dosomething,args=[i]) for i in range(1,11)]
for thread in threads:
     thread.start()

end_time=time()

print(f"main thread finishes at {end_time-start_time}")
```

```python
#  using thread but  waiting for child to be finished execution

from time import time,sleep
import threading

start_time=time()
def dosomething(id):

    print("Going to sleep...",id)
    sleep(1)
    print(f"Woken up...{id}")

threads=[threading.Thread(target=dosomething,args=[i]) for i in range(1,11)]
for thread in threads:
     thread.start()

for thread in threads:
     thread.join()

end_time=time()

print(f"main thread finishes at {end_time-start_time}")
```

```python
import threading
import time

balance = 0   # shared resource

def deposit(amount, times, lock):
    global balance
    for _ in range(times):
        lock.acquire()
        try:
            print(f"[{threading.current_thread().name}] Depositing: {amount}")
            current = balance
            time.sleep(0.1)   # simulate delay
            balance = current + amount
            print(f"[{threading.current_thread().name}] New Balance: {balance}")
        finally:
            lock.release()

def withdraw(amount, times, lock):
    global balance
    for _ in range(times):
        lock.acquire()
        try:
            print(f"[{threading.current_thread().name}] Withdrawing: {amount}")
            if balance >= amount:
                current = balance
                time.sleep(0.1)
                balance = current - amount
                print(f"[{threading.current_thread().name}] New Balance: {balance}")
            else:
                print(f"[{threading.current_thread().name}] Withdrawal Failed: Insufficient Funds")
        finally:
            lock.release()

# ------------------ MAIN ------------------

lock = threading.Lock()

t1 = threading.Thread(target=deposit, args=(100, 3, lock), name="DepositThread")
t2 = threading.Thread(target=withdraw, args=(80, 3, lock), name="WithdrawThread")

t1.start()
t2.start()

t1.join()
t2.join()

print("\nFinal Balance =", balance)
```
