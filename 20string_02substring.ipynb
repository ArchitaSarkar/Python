```python
# indexing 
# Strings are indexed, meaning each character has a specific position.
# two types of indexing: positive and negative.
# Positive indexing starts from 0 (left to right)
# Negative indexing starts from -1 (right to left)  
my_string = "Hello, World!"
print(my_string[0])    # H
print(my_string[-1])   # !
print(my_string[7])    # W
print(my_string[-5])   # o


# slicing
# we can extract a substring from a string using slicing.
print(my_string[0:5])  # Hello
print(my_string[7:12]) # World
print(my_string[:5])   # Hello
print(my_string[7:])   # World! 
print(my_string[:])    # Hello, World!
print(my_string[::2]) # Hlo ol!

#negative slicing
print(my_string[-6:-1]) # World  
print(my_string[-1:-6:-1]) #!dlro

#reverse a string using slicing
print(my_string[::-1])  # !dlroW ,olleH
```
