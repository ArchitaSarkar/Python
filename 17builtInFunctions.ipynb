```python
# some builtin functions are:
# print()
print("Hello, World!")

# input()
input("Enter something: ")

# type()
print(type("Hello"))

# int()
print(int("5"))

# str()
print(str(5))

# float()
print(float("5.5"))

# len()
print(len("Hello"))  # 5

# range()
for i in range(5):
    print(i)  # prints 0 to 4

# abs()
abs(-10)

# round()
print(round(5.67))  # 6
print(round(5.34))  # 5

# etc.

#pow
print(pow(2,3))  # 2^3 = 8

#min/max
print(min(2,3,1,4))  # 1
print(max(2,3,1,4))  # 4

#divmod
print(divmod(10,3))  # (3,1)  => quotient=3, remainder=1

#bin/oct/hex
print(bin(10))  # 0b1010    
print(oct(10))  # 0o12     
print(hex(10))  # 0xa

#sum
print(sum([1,2,3,4,5]))  # 15

#id
a=5
print(id(a))  # prints the memory address of variable a

#ord/chr
print(ord('A'))  # 65
print(chr(65))  # 'A'

#help
# help(print)  # gives help about print function



```
