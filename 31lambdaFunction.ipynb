ðŸ”¥ Difference Between Normal Function and Lambda Function

A normal function in Python is defined using the def keyword and usually has a name. It can contain multiple statements, including loops, conditions, print statements, or even other function calls. A normal function can also have a docstring for documentation and explicitly uses the return keyword to send back a value. Because it supports multiple lines and complex logic, normal functions are preferred when the code needs to be readable, reusable, and well-structured.

A lambda function, on the other hand, is an anonymous, one-line function created using the lambda keyword. It does not have a name unless you assign it to a variable. A lambda function is restricted to only one expression and cannot contain multiple statements like loops or print statements. It returns the value of that single expression automatically without using the return keyword. Lambda functions are mainly used for short, quick operations or when passing a function as an argument to other functions like map, filter, or sorted.

In simple terms, a normal function is suitable for larger, more complex logic, while a lambda function is used for small, temporary operations that can be written in a single line.

```python
# it can be used with higher order function

def sum(func,L):
    result=0
    for i in L:
        if func(i):
            result+=i
    return result

a=lambda x:x%2==0
b=lambda x:x%2!=0
c=lambda x:x%3==0

L=[1,2,3,4,5,6,7]
print(sum(a,L))
print(sum(b,L))
print(sum(c,L))

   
```

```python
# map
# filter
# reduce

L=[1,2,3,4,5,6]
print(list(map(lambda x:x*2,L)))
print(list(map(lambda x:"even" if x%2==0 else "odd",L
)))
```

```python
print(list(filter(lambda x:x>3,L)))
```

```python
# reduce
import functools
L1=[1,2,3,4,5,6,7]
print(functools.reduce(lambda x,y:x+y,L1)) # sum of list
print(functools.reduce(lambda x,y: x if x>y else y,L1)) # max element
print(functools.reduce(lambda x,y: x if x<y else y,L1)) # min element
```

```python
# list comprehension
L=[1,2,3,4,5,6,7]
L1=[item*2 for item in L]
print("L1=",L1)
L2=[i**2 for i in range(1,11)]
print("L2=",L2)
L3=[i*2 for i in range(1,11) if i%2==0]
print("L3=",L3)


```

```python
names = ["a", "b", "c"]
values = [10, 20, 30]

d = {names[i]: values[i] for i in range(len(names))}
print(d)
```
