```python
# literal is a fixed value assigned to a variable.
#there are 4 type of literals in python:
#1. Numeric literals    
#2. String literals
#3. Boolean literals
#4. Special literals (None)
```

```python
a=0b1100  #binary literal
b=100     #decimal literal  
c=0o234   #octal literal
d=0x12fc  #hexadecimal literal
print(a)
print(b)    
print(c)
print(d)
```

```python
x=3+8j #Numeric literal - complex
print(x,x.imag,x.real)
```

```python
a=True+4 # boolean literal
b=False+10
print(a)
print(b)
```

```python
a=None # special literal= special value representing absence of value
# best way to declare a variable with no value
print(a)
```
