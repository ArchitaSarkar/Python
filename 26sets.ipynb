```python
# sets do not allow duplicate values
# sets have no indexing or slicing
# sets do not allow mutable elements like lists or dictionaries
# set itself is a mutable data type

```

```python
s1={}
print(type(s1))  # this is a dictionary, not a set
```

```python
s1=set()
print(type(s1))  # this is a set
```

```python
s1={1,2,3,4,5,5,4,3} # homogeneous set
print(s1)
s2={1, 2.5, "hello", (1,2,3), 4+5j} # heterogeneous set
print(s2)
```

```python
# access
T1 = {10, 20, 30, 40, 50}
print(T1[0]) 
```

```python
# sets cannot be accessed by indexing or slicing
# sets can not be edited
# but we can add or remove elements
s1={10,20,30}
id1 = id(s1)
print(id1)
s1.add(40)
print(s1)
id2 = id(s1)
print(id2)
s1.remove(20)
print(s1)
id3 = id(s1)
print(id3)

if id1 == id2 == id3:
    print("set is mutable")
```

```python
# delete
# del
# remove
# pop
s1={10,20,30,40,50}
print(s1)
del s1
#print(s1)  # this will give error as s1 is deleted
s2={10,20,30,40,50}
print(s2)
s2.remove(30)
print(s2)
s3={10,20,30,40,50}
print(s3)
popped_element = s3.pop()
print(popped_element)
print(s3)
```
