```python
# dictionary has no indexing or slicing
# dictionary is mutable
# keys are immutable but values can be mutable or immutable
# keys should be unique


# mutable -> list, dictionary, set
# immutable -> int, float, string, tuple,boolean,complex

```

```python
D={}
D={'name':'archit','age':21,'marks':[90,95,88]}
print(D)
```

```python
D1={[1,2]:'archi'}
```

```python
D1={(1,2):'archi'}
```

```python
D1={1:2,1:3}
print(D1)
```

```python
D2={'name':'archit','age':21,'marks':[90,95,88]} # 2d dictionary
```

```python
# access
print(D2['name'])
print(D2.get('age'))
print(D2.get('marks')[1])
```

```python
# edit
D2['age']=22
print(D2)
```

```python
# add new key value pair
D2['gender']='male'
print(D2)   
```

```python
# delete key value pair
del D2['marks']
print(D2)
```

```python
# concatenation and repetition not possible
# membership operator
print('name' in D2)
print('marks' not in D2)
```

```python
# looping through dictionary
for key in D2:
    print(key,':',D2[key])
for key,value in D2.items():
    print(key,':',value)
```

```python
# functions
print(len(D2))
print(D2.keys())
print(D2.values())
print(D2.items())
```
