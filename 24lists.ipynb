```python
# list are used to store multiple items in a single variable.
# arrays are homogeneous data structures that store items of the same type. But lists can store items of different types.
# elements are stored in contiguous memory locations in arrays, while in lists, elements are stored as references to objects.
# arrays are much faster and more memory efficient for large datasets of the same type, while lists are more flexible and easier to use for general-purpose programming.
# Python provides a built-in list type, while arrays can be implemented using the array module or third-party libraries like NumPy.
```

```python
L1=[1,2,3] # homogeneous list
L2=[1,2,[1,2],"kumar","singh"] # heterogeneous list
print(L1)
print(L2)
```

```python
# list operations

# concatenation
L1=[1,2,3]
L2=[4,5,6]
L3=L1+L2
print(L3)
print(L1*3) # repetition
print(2 in L1) # membership
print(len(L1)) # length
print(max(L1)) # maximum
print(min(L1)) # minimum
print(sum(L1)) # sum
print(sorted(L1,reverse=True)) # sorting
print(reversed(L1)) # reverse iterator
print(list(reversed(L1))) # reversed list
```
