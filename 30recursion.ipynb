```python
def mult(x,y):
    if y==0:
      return 0
    else:
        return x+mult(x,y-1)

print(mult(5,8)) 
     
```

```python
def palindrome(s):
    if len(s) < 2:
        return True
    if s[0] != s[-1]:
        return False
    return palindrome(s[1:-1])

print(palindrome("racecar"))  # True
print(palindrome("hello"))    # False
```

```python
def fib(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib(n-1) + fib(n-2)

print(fib(13))  
```

```python
# memoization to optimize fibonacci
fib_cache = {}
def fib_memo(n):
    if n in fib_cache:
        return fib_cache[n]
    if n <= 0:
        value = 0
    elif n == 1:
        value = 1
    else:
        value = fib_memo(n-1) + fib_memo(n-2)
    fib_cache[n] = value
    return value
```
