```python
# parameter vs argument
# parameter is variable in the declaration of function
# argument is the actual value of this variable that gets passed to function
# types of arguments
# 1. positional arguments 
# example
def add(a,b):
    return a+b
print(add(2,3)) # 2 and 3 are positional arguments
# 2. keyword arguments
# example
def subtract(a,b):
    return a-b  
print(subtract(b=5,a=10)) # here we are using keyword arguments
# 3. default arguments
# example
def multiply(a,b=2):
    return a*b
print(multiply(5)) # here b will take default value 2# if we provide value for b then it will override the default value
print(multiply(5,3)) # here b will take value 3
# 4. variable length arguments
# example
def var_length(*args):
    return sum(args)
print(var_length(1,2,3,4,5)) # here args will be a tuple (1,2,3,4,5)
# 5. arbitrary keyword arguments
# example
def arb_keyword(**kwargs):
    return kwargs
print(arb_keyword(a=1,b=2,c=3)) # here kwargs will be a dictionary {'a':1,'b':2,'c':3}


```

```python
# there is a global variable 
# global variable can be accessed inside a function
# but can not be modified inside a function unless we use global keyword
x = 10
def func():
    global x
    x = x + 5
    return x
print(func()) # 15
print(x) # 15
```

```python
# nested functions
def outer_function(text):
    def inner_function():
        return text.upper()
    return inner_function()
print(outer_function("hello")) # HELLO
```

```python
# everything in python is an object including functions
# function as object
def greet(name):
    return "Hello " + name  
print(greet("Alice"))
print(type(greet)) # <class 'function'>
# we can assign function to a variable
g = greet
print(g("Bob")) # Hello Bob
# we can pass function as argument to another function
```

```python
del greet
print(greet("Alice")) # NameError: name 'greet' is not defined
```

```python
print(g("Charlie")) # Hello Charlie
```
