```python
# break means to stop the loop wherever we want
# continue means to skip the current iteration and move to next iteration of loop
# pass means to do nothing, it is used as a placeholder
```

```python
for i in range (1,11) :
    if i==7:
        break
    print(i)
```

```python
for i in range (1,11) :
    if i==7:
        continue
    print(i)
```

```python
for i in range(10) :
    pass
```
