```python
# Strings are sequences of characters used to store text data in Python.
# we can use single quotes, double quotes, or triple quotes to define strings.
# Example:
string1 = 'Hello, World!'
string2 = "Python is fun."
string3 = '''This is a multi-line 
string.'''  
print(string1)
print(string2)
print(string3)
```

```python
# we can create strings using the str() constructor
string4 = str(123)
print(string4)
print(type(string4))
```
