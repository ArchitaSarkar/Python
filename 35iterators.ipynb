✅ 1. Iterable

An Iterable is any Python object that can be looped over.

Examples of iterables:

list, tuple, set, dict, string

Objects that implement __iter__() method

✔ Example:
nums = [1, 2, 3]


nums is an iterable because you can do:

for n in nums:
    print(n)

Internally:

When you do…

iter(nums)


Python creates an iterator from the iterable.

✅ 2. Iterator

An Iterator is an object that produces values one at a time.

An iterator must implement:

__iter__() → returns itself

__next__() → returns next value
(and raises StopIteration when done)

✔ Example:
nums = [10, 20, 30]
it = iter(nums)      # iterator created

print(next(it))      # 10
print(next(it))      # 20
print(next(it))      # 30
print(next(it))      # StopIteration error

✅ 3. Iteration

Iteration means going through elements one-by-one.

Example:

for x in [1, 2, 3]:
    print(x)


The for loop automatically:

Converts iterable → iterator

Calls next() repeatedly

Stops when StopIteration is raised

```python
L=[i for i in range(1,100000000)]
import sys

print(sys.getsizeof(L))

x=range(1,100000000)
print(sys.getsizeof(x))
```
