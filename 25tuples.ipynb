```python
# tuples are immutable sequences, typically used to store collections of heterogeneous data
# they are defined by enclosing values in parentheses

# create
# access
# edit
# add
# delete
# operations
# functions
```

```python
T1=()
print(type(T1))
```

```python
T1 = (1, 2, 3, 4, 5) # homogeneous tuple
T2 = (1,2, 'c', 'd', 'e')  # heterogeneous tuple
print(T1)
print(T2)
```

```python
T3=(1)  # single element tuple
print(type(T3))
```

```python
T3=(1,)  # single element tuple
print(type(T3))
```

```python
tuple("hello")
```

```python
tuple([1,2,3,4,5]) # list to tuple conversion
```

```python
# access
T1 = (10, 20, 30, 40, 50)
print(T1[0])  # first element
print(T1[2])  # third element
print(T1[-1]) # last element
print(T1[-3]) # third last element
```

```python
# edit
# tuples are immutable, so we cannot edit them directly
# but we can convert them to a list, edit the list, and convert it back to a tuple
T1 = (10, 20, 30, 40, 50)
temp_list = list(T1)
temp_list[2] = 99
T1 = tuple(temp_list)   
print(T1)
```

```python
# delete
# tuples are immutable, so we cannot delete elements directly
# but we can convert them to a list, delete elements from the list, and convert it
# back to a tuple
T1 = (10, 20, 30, 40, 50)
temp_list = list(T1)
del temp_list[1]  # delete second element
T1 = tuple(temp_list)
print(T1)
# delete entire tuple
del T1
print(T1)  # this will raise an error since T1 is delete
```

```python
# operations
T1 = (1, 2, 3)
T2 = (4, 5, 6)
T3 = T1 + T2  # concatenation
print(T3)
T4 = T1 * 3   # repetition
print(T4)
# membership
print(2 in T1)  # True
print(5 not in T1)  # True
# slicing
T5 = (10, 20, 30, 40, 50)
print(T5[1:4])  # (20, 30, 40)
print(T5[:3])   # (10, 20, 30)
print(T5[2:])   # (30, 40, 50)
# length
print(len(T5))  # 5
# functions
T1 = (5, 2, 9, 1, 7)
print(min(T1))  # 1
print(max(T1))  # 9
print(sum(T1))  # 24
print(sorted(T1))  # [1, 2, 5, 7, 9]
print(T1.count(2))  # 1
print(T1.index(9))  # 2
print(T1)
```
